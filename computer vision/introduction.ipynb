```python
img_gray=cv2.imread(r'D:\DL projects\computer vision\cat.jpg',cv2.IMREAD_GRAYSCALE)
img_gray
```

```python
img_gray.shape
```

```python
cv2.imshow('CAT',img_gray)
cv2.waitKey()
cv2.destroyAllWindows()
```

```python
(thresh,binary_img)=cv2.threshold(img_gray,128,255,cv2.THRESH_BINARY)
# 0->128  black
# 128->255 white
print(thresh)
cv2.imshow('CAT binary',binary_img)
cv2.waitKey()
cv2.destroyAllWindows()
```

```python
cv2.imshow('Image',img)
cv2.imshow('CAT binary',binary_img)
cv2.imshow('CAT',img_gray)
cv2.waitKey()
cv2.destroyAllWindows()
```

```python
converted_image=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
plt.imshow(converted_image)
```

```python
cv2.rectangle(converted_image,pt1=(900,100),pt2=(1700,1000),color=(255,0,0),thickness=7)
plt.imshow(converted_image)
```

```python
converted_image=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
cv2.circle(converted_image,center=(1250,800),radius=225,color=(112,212,100),thickness=5)
plt.imshow(converted_image)
```

```python
#    puttext()
converted_image=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
cv2.putText(converted_image,text="cute cat",org=(250,1600),fontFace=cv2.FONT_HERSHEY_COMPLEX,fontScale=7,color=(0,0,255),thickness=5)
plt.imshow(converted_image)
```
