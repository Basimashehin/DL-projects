```python
import cv2
img=cv2.imread(r'D:\DL projects\eye detection\eye.jpg')
img
```

```python
import matplotlib.pyplot as plt
plt.imshow(img)
```

```python
con_img=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
plt.imshow(con_img)
```

```python
eye_cascade=cv2.CascadeClassifier(r'D:\DL projects\eye detection\haarcascade_eye.xml')
eye_cascade
```

```python
img_gray=cv2.cvtColor(con_img,cv2.COLOR_RGB2GRAY)
img_gray
```

```python
eyes=eye_cascade.detectMultiScale(img_gray,minNeighbors=1)
eyes
```

```python
len(eyes)
```

```python
for(x,y,w,h) in eyes:
    cv2.rectangle(con_img,(x,y),(x+w,y+h),(0,0,255),4)
```

```python
video=cv2.VideoCapture(r'D:\DL projects\face detection\face video.mp4')
while True:
    sucess,frame=video.read()
    gray_img=cv2.cvtColor(frame,cv2.COLOR_BGR2GRAY)
    eyes=eye_cascade.detectMultiScale(gray_img,minNeighbors=32)
    # print(len(eyes))
    for(x,y,w,h) in eyes:
        cv2.rectangle(frame,(x,y),(x+w,y+h),(0,0,255),4)
    cv2.imshow('video',frame)
    if cv2.waitKey(1) & 0XFF==ord('q'):
        break
# video.release()
cv2.destroyAllWindows()
```

```python
video=cv2.VideoCapture(0)
while True:
    sucess,frame=video.read()
    gray_img=cv2.cvtColor(frame,cv2.COLOR_BGR2GRAY)
    eyes=eye_cascade.detectMultiScale(gray_img,minNeighbors=14)
    for(x,y,w,h) in eyes:
        cv2.rectangle(frame,(x,y),(x+w,y+h),(0,255,255),4)
    cv2.imshow('webcam',frame)
    if cv2.waitKey(1) & 0XFF==ord('q'):
        break
video.release()
cv2.destroyAllWindows()

    
```
